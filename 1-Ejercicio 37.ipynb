```python
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
```

```python
sqlContext = SQLContext(sc)
```

```python
df = sqlContext.read.parquet('/content/drive/MyDrive/2021-2C/ORGANIZACION DE DATOS (7506)/TPs/TP2/Wikipedia_dataset-2c2021/redirect_list.parquet', header=True, inferSchema=True)
df
```

```python
rdd = df.rdd
rdd.take(5)
```

```python
rdd.count()
```

```python
rd_appearances = rdd.map(lambda x: (x.rd_title,1)).reduceByKey(lambda x,y: x+y).takeOrdered(1, lambda x: -x[1])
```

```python
rd_appearances
```

# Respuesta

```python
print("El contenido al que más se hacen redirecciones es", rd_appearances[0][0])
```
