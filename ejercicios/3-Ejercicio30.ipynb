```python
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials
from pyspark.sql import *
from pyspark.sql.functions import *
from pyspark import SparkContext
from pyspark.sql import SQLContext
import pandas as pd
```

```python
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
```

```python
sqlContext = SQLContext(sc)
```

```python
df_geotags = sqlContext.read.parquet('/content/drive/MyDrive/2021-2C/ORGANIZACION DE DATOS (7506)/TPs/TP2/Wikipedia_dataset-2c2021/geo_tags.parquet', header=True, inferSchema=True)
```

```python
df_geotags
```

```python
rdd_geotags = df_geotags.rdd
```

```python
rdd_geotags.take(5)
```

```python
earth_content = rdd_geotags.filter(lambda x: x.gt_globe == 'earth').count() / rdd_geotags.count()
earth_content
```

```python
non_earth_content = 1 - earth_content
non_earth_content
```

# Respuesta


```python
print("El porcentaje de contenido geolocalizado de wikipedia NO está en la tierra es",non_earth_content*100)
```
